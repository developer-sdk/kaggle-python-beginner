# 정렬
인덱스, 칼럼 기준으로 데이터를 정렬합니다. 

+ 정렬 함수
    + sort_index(): 인덱스를 정렬
    + sort_values(): 칼럼값을 기준으로 정렬
+ 정렬 함수 속성 
    + 정렬 기준: axis
        + 0: 인덱스 기준
        + 1: 칼럼 기준
    + 정렬 방향: ascending
        + True: 오름차순
        + False: 내림차순

## 인덱스 정렬 
`sort_index()`는 인덱스를 이용한 데이터의 정렬입니다. 

```python
train
```

```python
# 인덱스를 기준으로 내림차순으로 정렬(인덱스가 890 -> 889 -> 888 순으로 나타남)
train.sort_index(axis=0, ascending=False)
```

```python
# 칼럼을 기준으로 내림차순으로 정렬
train.sort_index(axis=1, ascending=False)
```

```python
# 칼럼을 기준으로 오름차순 정렬 
train.sort_index(axis=1, ascending=True)
```

# 결측치
결측치는 `isnull()` 함수를 이용하여 확인할 수 있습니다. 결측치가 아닌 값을 확인할 때는 `notnull()`을 이용합니다. 

```python
# 결측치를 더하여 건수로 확인 
train.isnull().sum()
```

```python
# 결측치가 아닌 값을 더하여 건수로 확인 
train.notnull().sum()
```

### 결측치 채우기 
결측치를 다른 값으로 치환할 때는 `fillna()`를 이용합니다. 데이터프레임의 `fillna()`를 이용하여 값을 채울때는 딕셔너리를 전달합니다. 

```python
train.Age.fillna(-1)
```

# 그룹핑 
그룹핑 하여 데이터를 구성할 때는 `groupby()`를 이용합니다. 그룹핑 하고 난후 통계함수나 `agg()`를 이용하여 통계정보를 구할 수 있습니다. 

```python
# Pclass, Survived를 조회하여, Pclass를 그룹핑하여 Survived의 빈도 계산 
train[['Pclass', 'Survived']].groupby(['Pclass']).count()
```

```python
# as_index=False로 주면 그룹핑값을 인덱스로 하지 않고 값으로 반환 
train[['Pclass', 'Survived']].groupby(['Pclass'], as_index=False).sum()
```

### agg: 두 개의 결과 동시 확인 
`agg`는 집계 함수를 전달하여 여러 개의 결과를 동시에 확인할 수 있습니다. 

```python
train[['Pclass', 'Survived']].groupby(['Pclass']).agg({'Survived': ['count', 'sum']})
```

# for 문을 이용한 row 반복 처리
`iterrows()` 함수를 이용하여 반복문을 처리할 수 있습니다. 반복문에 index와 Series타입의 row가 전달됩니다. 

```python
for index, row in train.iterrows():
    print("index: {0}".format(index))
    print("--------------")
    print(row)
    print("--------------")
    print(row['PassengerId'])
    break
```

# SQL 처리 
데이터프레임을 테이블로 설정하고 SQL 문을 이용하여 처리할 수 있습니다. 

```python
from sqlalchemy import create_engine

# sqlite 엔진을 이용하여 데이터프레임을 테이블로 변환 
engine = create_engine('sqlite://', echo=False)
train.to_sql('train', con=engine)
engine.execute("SELECT * FROM train").fetchall()
engine.execute("""SELECT Sex, Survived, count() 
                    FROM train 
                   Group by Sex, Survived""").fetchall()
```
