# CSV
### 읽기 
CSV 파일을 읽을 때는 `read_csv`를 이용합니다. 로컬의 파일을 읽을 수 있고, HTTP주소를 전달하여 외부의 파일을 읽을 수도 있습니다. 

```python
import pandas as pd

# 로컬의 파일을 읽을 때 
train = pd.read_csv(r'/Users/seo/kaggle/titanic/train.csv')
train.head(3)
```

```python
# http로 접근 가능 
http_url = "https://raw.githubusercontent.com/developer-sdk/kaggle-python-beginner/master/datas/kaggle-titanic/train.csv"
train = pd.read_csv(http_url)
train.head(3)
```

### 쓰기 
CSV 파일을 쓸 때는 `to_csv`를 이용합니다. 

```python
train.to_csv('output.csv', index=False)
```
